```python
from pylab import *
from pandas import *
from seaborn import *
```

```python
dt=read_csv('advertising.csv');dt.head()
```

```python
dt.Male.replace([0,1],["woman","man"],inplace=True)
```

```python
dt.head()
```

```python
pairplot(dt,hue='Male')
```

```python
dt['Clicked on Ad'].replace([0,1],["No","Yes"],inplace=True)
```

```python
dt.columns
```

```python
pairplot(dt,hue='Clicked on Ad')
```

```python
X=dt[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']]
y=dt['Clicked on Ad']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(X,y,random_state=0)
model=LogisticRegression().fit(x_tr,y_tr)
```

```python
print('train core: {:.3f}'.format(model.score(x_tr,y_tr)))
print('text core: {:.3f}'.format(model.score(x_te,y_te)))
```
